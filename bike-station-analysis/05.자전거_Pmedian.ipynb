```python
import pandas as pd
import pulp
from math import radians, sin, cos, sqrt, atan2
```

```python
incheon_df = pd.read_csv("/content/4.5km이내_시설위치_관광지포함.csv")
incheon_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
incheon_df.head()
```

```python
def solve_p_median(incheon_df, p):
    incheon_df['distance_km'] = pd.to_numeric(incheon_df['distance_km'], errors='coerce')

    locations = incheon_df['장소명'].unique()
    prob = pulp.LpProblem('p_median', pulp.LpMinimize)

    x = pulp.LpVariable.dicts('x', [(o, d) for o in locations for d in locations], 0, 1, pulp.LpBinary)
    y = pulp.LpVariable.dicts('y', locations, 0, 1, pulp.LpBinary)

    prob += pulp.lpSum(incheon_df.loc[(incheon_df['장소명'] == o) & (incheon_df['장소명'] == d), 'distance_km'].values[0] * x[(o, d)]
                       for o in locations for d in locations if not incheon_df.loc[(incheon_df['장소명'] == o) & (incheon_df['장소명'] == d), 'distance_km'].empty)

    for origin in locations:
        prob += pulp.lpSum(x[(origin, destination)] for destination in locations) == 1
        for destination in locations:
            prob += x[(origin, destination)] <= y[destination]

    prob += pulp.lpSum(y[location] for location in locations) == p

    prob.solve()

    medians = [location for location in locations if pulp.value(y[location]) == 1]
    return medians
```

```python
medians = solve_p_median(incheon_df, 20)
median_df = incheon_df[incheon_df['장소명'].isin(medians)][['장소명', '구분', 'distance_km', 'lon', 'lat']]

median_df.to_csv("인류미포함pmedian.csv")
```

```python
def solve_p_median(incheon_df, p):

    incheon_df['distance_km'] = pd.to_numeric(incheon_df['distance_km'], errors='coerce')
    incheon_df['stay_cnts'] = pd.to_numeric(incheon_df['stay_cnts'], errors='coerce')

    locations = incheon_df['장소명'].unique()

    prob = pulp.LpProblem('p_median', pulp.LpMinimize)

    x = pulp.LpVariable.dicts('x', [(o, d) for o in locations for d in locations], 0, 1, pulp.LpBinary)
    y = pulp.LpVariable.dicts('y', locations, 0, 1, pulp.LpBinary)

    prob += pulp.lpSum(incheon_df.loc[(incheon_df['장소명'] == o) & (incheon_df['장소명'] == d), 'distance_km'].values[0] *
                       incheon_df.loc[incheon_df['장소명'] == o, 'stay_cnts'].values[0] * x[(o, d)]
                       for o in locations for d in locations if not incheon_df.loc[(incheon_df['장소명'] == o) & (incheon_df['장소명'] == d), 'distance_km'].empty)
    for origin in locations:
        prob += pulp.lpSum(x[(origin, destination)] for destination in locations) == 1
        for destination in locations:
            prob += x[(origin, destination)] <= y[destination]

    prob += pulp.lpSum(y[location] for location in locations) == p

    prob.solve()

    medians = [location for location in locations if pulp.value(y[location]) == 1]
    return medians

medians = solve_p_median(incheon_df, 20)
median_df = incheon_df[incheon_df['장소명'].isin(medians)][['장소명', '구분', 'distance_km', 'lon', 'lat', 'stay_cnts']]

median_df
```

### MCLP

```python
import pulp
import numpy as np
from geopy.distance import geodesic

def solve_mclp(data_df, festi_loc, max_facilities, d_max):
    facilities = data_df['장소명'].unique()
    demands = festi_loc['축제명'].unique()

    prob = pulp.LpProblem('MCLP', pulp.LpMaximize)

    x = pulp.LpVariable.dicts('x', facilities, 0, 1, pulp.LpBinary)
    y = pulp.LpVariable.dicts('y', (demands, facilities), 0, 1, pulp.LpBinary)
    def calculate_distance(festi_row, facility_row):
        return geodesic((festi_row['lat'], festi_row['lon']), (facility_row['lat'], facility_row['lon'])).km

    prob += pulp.lpSum(y[demand][facility] for demand in demands for facility in facilities)

    for demand in demands:
        prob += pulp.lpSum(y[demand][facility] for facility in facilities) <= 1

    prob += pulp.lpSum(x[facility] for facility in facilities) <= max_facilities

    for demand in demands:
        for facility in facilities:
            festi_row = festi_loc[festi_loc['축제명'] == demand].iloc[0]
            facility_row = data_df[data_df['장소명'] == facility].iloc[0]
            distance = calculate_distance(festi_row, facility_row)

            if distance <= d_max:
                prob += y[demand][facility] <= x[facility]
            else:
                prob += y[demand][facility] == 0

    prob.solve()


    selected_facilities = [facility for facility in facilities if x[facility].value() == 1]
    return selected_facilities
```
