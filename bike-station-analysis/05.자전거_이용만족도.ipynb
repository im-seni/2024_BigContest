```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
```

```python
satis = pd.read_csv("/content/인천광역시_자전거 이용 환경 만족도_20211231.csv", encoding='cp949')
satis.info()
```

```python
satis[satis['특성별(1)']=='군구별']
```

```python
manjok = satis[satis['특성별(1)']=='군구별'].fillna(0)
manjok.head()
```

```python
satis['특성별(1)'].value_counts()
```

```python
manjok.columns = manjok.columns.str.replace(' (퍼센트)', '')

def plot_donut_chart(dataframe, index):
    labels = ['매우 불만족', '약간 불만족', '보통', '약간 만족', '매우 만족']
    values = dataframe.loc[index, labels].values
    특성 = dataframe.loc[index, '특성별(2)']

    colors = ['#001f3f', '#0074D9', '#7FDBFF', '#E2E2E2', '#0000FF']

    fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.5, marker_colors=colors)])
    fig.update_layout(
        title_text=f'{특성} 자전거 이용환경 만족',
        annotations=[{'text': 특성, 'x': 0.5, 'y': 0.5, 'font_size': 20, 'showarrow': False}]
    )
    fig.show()
```

```python
for i in range(len(manjok)):
    plot_donut_chart(manjok, i)
```

```python
fig = px.bar(manjok, x='특성별(2)', y='5점 평균 (점)', title='5점 평균 바차트',
             labels={'특성별(2)': '지역', '5점 평균 (점)': '평균 점수'},
             text='5점 평균 (점)')

fig.update_traces(marker_color='#0074D9', textposition='outside')

fig.update_yaxes(range=[0, 6], tickvals=[0, 1, 2, 3, 4, 5])

converted_score = 4.845
fig.add_shape(
    type="line",
    line_color="red",
    line_width=2,
    opacity=1,
    x0=-0.5,
    x1=len(manjok) - 0.5,
    y0=converted_score,
    y1=converted_score,
)

fig.show()
```
